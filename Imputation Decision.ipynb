```python
drought = pd.read_csv("drought.csv")
storm = pd.read_csv("storm.csv")
wildfire = pd.read_csv("wildfire.csv")
extreme = pd.read_csv("extreme_temp.csv")
glacial = pd.read_csv("glacial.csv")
landslide = pd.read_csv("landslide.csv")
flood = pd.read_csv("flood.csv")

drought = drought[(drought.Year>=1970)&(drought.Year<=2020)]
drought.reset_index(drop=True, inplace=True)
storm = storm[(storm.Year>=1970)&(storm.Year<=2020)]
storm.reset_index(drop=True, inplace=True)
wildfire = wildfire[(wildfire.Year>=1970)&(wildfire.Year<=2020)]
wildfire.reset_index(drop=True, inplace=True)
extreme = extreme[(extreme.Year>=1970)&(extreme.Year<=2020)]
extreme.reset_index(drop=True, inplace=True)
glacial = glacial[(glacial.Year>=1970)&(glacial.Year<=2020)]
glacial.reset_index(drop=True, inplace=True)
landslide = landslide[(landslide.Year>=1970)&(landslide.Year<=2020)]
landslide.reset_index(drop=True, inplace=True)
flood = flood[(flood.Year>=1970)&(flood.Year<=2020)]
flood.reset_index(drop=True, inplace=True)
```

```python
#drought
grouped = drought.groupby("Country")

def percentages(df):
    a = df["Total.Affected"].notna().sum()
    percent = a / len(df)
    return percent

percents = grouped.apply(percentages)

percent = percents.to_frame().reset_index()
percent.columns = ["Country", "complete"]

enough = percent[percent["complete"] <= 0.5]

#for i in enough.Country:
    #print(i)
    
len(enough.Country)

for i in enough.Country:
    print(f'"{i}",')
```
