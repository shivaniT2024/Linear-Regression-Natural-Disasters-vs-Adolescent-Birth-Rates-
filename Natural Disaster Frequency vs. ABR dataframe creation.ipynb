```python
new = frequencymerged.groupby("Country")

def count(x):
    den = x['Adolescent'].count()
    num = x['gov_expenditure.csv'].count()
    percent = 0
    if den>0:
        percent = num/den
    return percent

counts = new.apply(count)

new = pd.DataFrame(counts)
new.columns = ["one"]
newer = new[new["one"]>=0.5]
newer.index
```
